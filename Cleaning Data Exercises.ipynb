## Missing Data

In the Towards Data Science article, they list 3 techniques for detecting any missing data frame. Try all three before deciding on a way to solve any missing data issues.

```python
# Plot a heatmap of any missing data
cols = df.columns[:30]
colours = ['#000099', '#ffff00'] #yellow = missing; blue = not missing
sns.heatmap(df[cols].isnull(),cmap=sns.color_palette(colours))
```

```python
# Try out a missing data percentage list! 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100))) # % = amount of data missing
```

```python
# Plot a missing data histogram
#creating missing indicator for features with missing data
for col in df.columns:
    missing = df[col].isnull()
    num_missing = np.sum(missing)
    
    if num_missing > 0:
        print('created missing indicator for: {}'.format(col))
        df['{}_ismissing'.format(col)] = missing
        
#based on indicator, plot histogram of missing values
ismissing_cols = [col for col in df.columns if 'ismissing' in col]
df['num_missing'] = df[ismissing_cols].sum(axis=1)

df['num_missing'].value_counts().reset_index().sort_values(by='index').plot.bar(x='index', y='num_missing')
```

Now that you have tried all 3 techniques for detecting missing data, did you find any? Reading through the article, which solution do you think will help you clean the data set and handle the missing data? Try out your solution below!

```python
# Handle any missing data in your dataframe.
# dropping features with high value percentage
cols_to_drop = ['Facility Type', 'Landmark', 'School or Citywide Complaint', 'Vehicle Type', 'Taxi Company Borough', 'Taxi Pick Up Location', 'Bridge Highway Name', 'Bridge Highway Direction', 'Road Ramp', 'Bridge Highway Segment', 'Garage Lot Name', 'Ferry Direction', 'Ferry Terminal Name', 'Intersection Street 1', 'Intersection Street 2']
df_less_cols = df.drop(cols_to_drop, axis=1)
df_less_cols
```

```python
# impute the missing data
#imputing missing values and creat missing value indicator variables for eavh non-numeric column
df_less_cols_non_numeric = df_less_cols.select_dtypes(exclude=[np.number])
non_numeric_cols = df_less_cols_non_numeric.columns.values

for col in non_numeric_cols:
    missing = df_less_cols[col].isnull()
    num_missing = np.sum(missing)
    
    if num_missing > 0: #only do imputation for the columns that have missing values
        print('imputing missing values for: {}'.format(col))
        df['{}_ismissing'.format(col)] = missing
        
        top = df_less_cols[col].describe()['top'] #impute with the most frequent value
        df_less_cols[col] = df_less_cols[col].fillna(top)
```
